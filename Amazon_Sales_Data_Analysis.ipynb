# As we see, we have 1465 rows and 16 columns in our Amzon Sales Dataset
# we have to convert the data type of  columns like "discounted_price" , "actual_price" , "discount_percentage" ,  "rating" , "rating_count"  into float or integer for better Analysis.

```python
# converting discounted_price datatype into Float
df["discounted_price"] = df["discounted_price"].str.replace('₹','', regex=False)
df['discounted_price'] = df["discounted_price"].str.replace(',','')
df["discounted_price"] = df["discounted_price"].astype(float)
```

```python
df["discounted_price"].dtype
```

```python
# Converting Actual_price datatype into Float
df["actual_price"] = df["actual_price"].str.replace('₹','')
df["actual_price"] = df["actual_price"].str.replace(',','')
df["actual_price"] = df["actual_price"].astype(float)
```

```python
df["actual_price"].dtype
```

```python
# converting discount_percentage datatype into integer 
df["discount_percentage"] = df["discount_percentage"].str.replace('%','').astype(float)
df["discount_percentage"] =df["discount_percentage"].astype(int)
```

```python
df["discount_percentage"].dtype
```

```python
# converting rating datatype into integer
df["rating"] = df["rating"].replace('|',np.nan)
df["rating"]=df["rating"].astype(float)
df["rating"]=df["rating"].fillna(0)
```

```python
df["rating"].dtype
```

```python
# converting rating_count
df["rating_count"] = df["rating_count"].str.replace(',','').fillna(0)
df["rating_count"] = df["rating_count"].astype(int)
```

```python
df['rating_count'].dtype
```

```python
df.describe()
```

```python
df["category"].describe
```

# we have category column in our dataset in which we have maily two category Computer Accessories and the Homes&kitchen so we have to extract the single name for these category for better data analysis

```python
#  first we have to extrat the home kitchen category due to dis-similirity in name 
df["category"] = df["category"].str.extract(r"(Home&Kitchen)")
```

```python
df["category"].describe
```

```python
# fill NaN values to  Computer Accessories 
df["category"] = df["category"].fillna("Computer Accessories")
```

```python
df["category"].describe
```

# Now Removing the unnecessary columns like img_link, about_product", review_title , review_content , product_name,user_id ,user_name ,review_id  and product_link for better data Analysis

```python
df.drop(["img_link", "about_product", "review_title" , "review_content" , "product_name","user_id" ,"user_name" ,"review_id","product_link"],axis=1 , inplace=True)
```
